### Processing Data
* First & last name: reduce all to lowercase
* Middle name: reduce to one uppercase first letter
* Sex: reduce all to one letter F or M
* street: reduce all to uppercase with street abbrevations, remove all hashtags
* street2: if one does not have letters, then take letters off of other one.
* city: reduce all to lower case
* state: reduce all to abbreviation using dictionary



```python
# dictionary for street abbreviations

d = { 
    'aly' : 'Alley',
    'ave' : 'Avenue',
    'blv' : 'Boulevard',
    'blvd' : 'Boulevard',
    'cir' : 'Circle',
    'ct' : 'Court',
    'cv' : 'Cove',
    'cyn' : 'Canyon',
    'dr' : 'Drive',
    'expy' : 'Expressway',
    'hwy' : 'Highway',
    'ln' : 'Lane',
    'pkwy' : 'Parkway',
    'pl' : 'Place',
    'pt' : 'Point',
    'rd' : 'Road',
    'sq' : 'Square',
    'st' : 'Street',
    'ter' : 'Terrace',
    'tr' : 'Trail',
    'trl' : 'Trail',
    'wy' : 'Way'
    }

streetDict = {}
for key, val in d.items():
    val = val.lower()
    streetDict[val] = key

print(streetDict)
```

```python
# functions for data processing
def name(s1):
    s1 = s1.lower()
    return s1

def middleNameandSex(s1):
    s2 = s1[0:1]
    s2 = s2.upper()
    return s2

def street(s1):
    for key, val in streetDict.items():
        s1 = s1.replace(key, val)
    s1 = s1.replace("#","")
    s1 = s1.lower()
    return s1
    
def street2(s1):
    # remove all letters
    letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    for letter in letters:
        s1 = s1.replace(letter, "")
        s1 = s1.replace(letter.upper(), "")
    s1 = s1.replace("#","")
    return s1
    
def city(s1):
    return s1.lower()

def state(s1):
    s1 = s1.upper()
    if len(s1)==2:
        return s1
    else:
        if (not s1 in stateDict):
            for key in stateDict.keys():
                if s1 in key:
                    s1 = key
                    abbrev = stateDict[s1]
                    return abbrev
                return 0
        else:
            abbrev = stateDict[s1]
            return abbrev
```

```python
def processData(df):
    for x in range(len(df)):
        for y in range(len(df.columns)):
            if df.iloc[x,y]==0:
                continue
            elif y in [1,3,11,13]:
                df.iloc[x,y] = name(df.iloc[x,y])
            elif y in [2,5,12]:
                df.iloc[x,y] = middleNameandSex(df.iloc[x,y])
            elif y in [6,14]:
                df.iloc[x,y] = street(df.iloc[x,y])
            elif y in [7,15]:
                df.iloc[x,y] = street2(df.iloc[x,y])
            elif y in [8,16]:
                df.iloc[x,y] = city(df.iloc[x,y])
            elif y in [9,17]:
                df.iloc[x,y] = state(df.iloc[x,y])
    return df
```

```python
df_processed = processData(df_test)
df_processed.head()
```
